# Data visualization with Seaborn - Part 2

The dataset which we work upon has many columns with datatypes as numerical, categorical or object type. In this article we will focus on plotting categorical data using seaborn package. For this, lets import the Seaborn package and load the built in dataset from seaborn.

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Barplot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

For barplot the 'x' should be categorical column and 'y' should be numerical column.
The above barplot shows us the average/mean of the total bill per categorical column value. As one can see the average total bill for male is slightly higher than average total bill for female. 

We can even pass a parameter called as 'estimator' which is basically a Statistical function to estimate within each categorical bin. By default it is set to average. We can also put other aggregate function. Here I will pass standard deviation as an estimator function.

```python
import numpy as np

#this will tell me what is the standard deviation of the total_bill per categorical column
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Countplot

This is similar to barplot except the estimator is counting the number of occurances. And because of that we only set the 'x' value and no 'y' value.

```python
sns.countplot(x='sex',data=tips)
```

### Boxplot

This type of plot are used to show the distribution of categorical data. Lets plot the graph first.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

This boxplot shows the quartiles of the dataset. The viscous shows the extent of the data. The points which are outside of the viscous are outliers. We can add 'hue' argument in the boxplot to distinguish it based on some category. Lets look at the below example.

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

The above plot is splitted per day and also based on the smoking habits of persons. This is the power of seaborn which allows you to add 'hue' argument to visualize the data categorically.

### Violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

As you can see the violin plot gives the distribution of the data as well.

### Strip plot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True, hue='sex')
```

### Swarm plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

Swarm plot is nothing but the combination of strip plot and violin plot. Sometimes this plot does not scale for large numbers. So avoid using swarm plots for large datasets.

### Factor plot

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

Factor plot is the one by which we can plot any graph by specifying 'kind' argument. We can pass the type of plot we want as an argument. Above we passed 'bar' for barplot. Below is the example of violin plot.

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
#count plot using factor plot
sns.factorplot(x='day',data=tips,kind='count')
```
