### Figure size, Aspect ratio and DPI

Matplotlib allows you to control each of the aspects and you can specify those while calling the object. Lets see an example.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0,5,11)
y = x** 2
```

```python
#figure() takes 2 arguments namely a tuple for size of the figure and dpi i.e. dots per inch.
fig = plt.figure(figsize=(8,2),dpi=100)

#add axes
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

**Multiple plots**

```python
fig,axes = plt.subplots(nrows = 2, ncols =1 , figsize=(6,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

#to avoid overlap
plt.tight_layout()
```

**Legends**

Consider below code where we are plotting 2 lines in the same canvas. But we dont know which plot is what. For this we can use legends.

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x , x **2, label = 'X Squared')
ax.plot(x, x ** 3, label = 'XCubed')

#add legend
ax.legend(loc=0) #loc is the location of the legend in the graph
```

### Setting colors, linewidths and linetypes

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#if linewidth is assigned with greater number you will get a thick line
#alpha tells you how transparent your line will be
#linestyle defines style of the line such as dashed or dotted
#markers will tell us where exactly our values points on the axes

ax.plot(y,x,color='red', linewidth = 2, linestyle= ':',marker = '+')
ax.plot(x,y,color='purple', linewidth = 5, alpha = 0.5, linestyle= '-.', marker='o',markersize=8, markerfacecolor='yellow') # colors accepts basic color strings or RGB hex code
```

**How to set x limit and y limit**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='blue', linewidth = 2, linestyle='--')

#set limits
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
