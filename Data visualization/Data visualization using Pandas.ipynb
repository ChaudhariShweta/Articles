# Data visualization using Pandas

Visualization of the data can be achieved by using built in functions of Pandas library. Below are few examples of various plots we can see using this. 

```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#create a dummy dataframe
df = pd.DataFrame(np.random.randint(low=0, high=10, size=(5, 5)),columns=['a', 'b', 'c', 'd', 'e'])
```

```python
#check the head
df.head()
```

```python
#histogram for any column
df['a'].hist(bins=25)
```

```python
#plot any graph using .plot() function and specifying 'kind' argument

df['b'].plot(kind='hist',bins=15)
```

```python
#directly call .hist() function on any column

df['c'].plot.hist()
```

```python
#function .area() is applied on entire dataframe

df.plot.area(alpha=0.4)
```

```python
#entire dataframe can be viewed using .bar() function

df.plot.bar(stacked = True)
```

```python
#line plot by specifying x and y values

df.plot.line(x=df.index, y='b')
```

```python
#scatter plot to compare any 2 columns

df.plot.scatter(x='a',y='b')
```

```python
#boxplot using .box() function

df.plot.box()
```

```python
#kernel density estimation i.e. kde plot for a dataframe

df.plot.kde()
```
