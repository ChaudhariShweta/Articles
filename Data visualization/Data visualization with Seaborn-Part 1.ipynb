### Get the data for plotting

```python
#import seaborn first
import seaborn as sns

#this is done to see visualization inside the notebook
%matplotlib inline   
```

Now lets get some data which we can pass to Seaborn functions for visualization. Seaborn comes with few built in datasets. Lets load them.

```python
tips = sns.load_dataset('tips')
```

```python
#check head() of tips dataset which tells people who had meal at the hotel and how much tip they left behind
tips.head()
```

### Distribution plots

Distplot allows you to visualize distribution by plotting a histogram.

```python
#distplot() function takes single column
sns.distplot(tips['total_bill'],bins=40)
```

By looking at the plot we can tell that most of our bills were between 10 and 20 dollars. Usually bin size depends on your dataset. If we have bin values very high then we may end up plotting every single instance of the data.

### Joint plots

Joint plot allows you to basicaly match up two distplots for bivariate data. Meaning you can combine two distplots. We have a parameter called as 'kind' which allows us to compare these two plots.

```python
#use .jointplot() to acheive this
sns.jointplot(x='total_bill',y='tip',data=tips)
```

The above jointplot tells us that we have Total_bill on x-axis and tip on y-axis and in between we have plotted scatter plot. The scatter plot tells us that we have direct relationship between these two variables and as you go for higher bill your tip amount will increase.

```python
#kind parameter helps us in comparing the plots in various ways. By default we will get scatter plot if we dont specify any kind
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#another kind is regression where in seaborn will plot a scatter plot along with a regression line
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

### Pair plot

Pair plot is essentially going to plot pairwise relationship across an entire dataframe only for numerical columns. It also supports 'hue' argument for categorical columns. There is one more argument called as 'palette' which will allow us to choose various colors for our graphs.

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

### Rug Plot

Rug plot takes single column as an argument just like distplot. Rug plot just draws a dashed mark for every point on univariate distribution i.e. essentially one variable.

```python
sns.rugplot(tips['total_bill'])
```
